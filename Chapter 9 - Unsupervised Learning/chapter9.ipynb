# Topic 2 — K-Means Clustering (Core Algorithm)

## 📍 Where to Read
**9_Unsupervised Learning Techniques.pdf**
- Section: K-Means

> K-Means is the most famous clustering algorithm and one of the most commonly used unsupervised methods.

---

## 1. What is Clustering?

$$\text{Group similar data points together}$$

- Points in the **same cluster** → very similar
- Points in **different clusters** → very different

**Example — customer spending behavior:**
```
Cluster 1 → budget customers
Cluster 2 → premium customers
Cluster 3 → medium spenders
```

---

## 2. What is K-Means?

K-Means divides a dataset into **K clusters**, where each cluster is represented by a **centroid**.

$$K = \text{number of clusters (chosen by you)}$$

---

## 3. What is a Centroid?

A centroid is the **center of a cluster** — mathematically, the mean of all points in it.

**Example:**
```
Cluster points: (2,3), (3,4), (4,5)
Centroid:       (3, 4)
```

---

## 4. K-Means Algorithm (Step-by-Step)

**Step 1 — Choose K:**
$$K = 3 \text{ clusters}$$

**Step 2 — Initialize centroids:** randomly pick K points as starting centroids (C1, C2, C3).

**Step 3 — Assign points to nearest centroid** using Euclidean distance:
$$\text{each point} \rightarrow \text{nearest centroid}$$

**Step 4 — Update centroids:**
$$\text{new centroid} = \text{mean(all points in cluster)}$$

**Step 5 — Repeat** steps 3 and 4 until clusters stop changing (**convergence**).

---

## 5. Code Example
```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=300, centers=4, random_state=42)

kmeans = KMeans(n_clusters=4, random_state=42)
y_pred = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.scatter(kmeans.cluster_centers_[:, 0],
            kmeans.cluster_centers_[:, 1],
            c='red', marker='X', s=200)
plt.title("K-Means Clustering")
plt.show()
```

> Red **X** marks the centroid of each cluster.

---

## 6. Important Properties

| ✅ Strengths | ⚠️ Assumptions |
|-------------|----------------|
| Simple & fast | Clusters are roughly **spherical** |
| Scalable to large datasets | Clusters have **similar sizes** |

---

## 🔑 Key Summary

**Goal:** Partition dataset into K clusters.

| Concept | Meaning |
|---------|---------|
| **Cluster** | Group of similar points |
| **Centroid** | Mean center of a cluster |
| **Euclidean distance** | Measure used to assign points |

**Algorithm:**
1. Initialize K centroids randomly
2. Assign each point to its nearest centroid
3. Update centroids as the mean of their cluster
4. Repeat until convergence

# Topic 3 — K-Means Optimization Objective (Cost Function)

## 📍 Where to Read
**9_Unsupervised Learning Techniques.pdf**
- K-Means section explaining the optimization objective
- Part discussing inertia

> Understanding this explains what K-Means is actually optimizing.

---

## 1. What K-Means is Trying to Minimize

K-Means minimizes a quantity called **inertia**, also known as:

$$\text{Within-Cluster Sum of Squared Distances (WCSS)}$$

---

## 2. Inertia — Core Concept

**Inertia** measures how close points are to their cluster centroid:

$$\text{Inertia} = \sum \| x_i - \text{centroid} \|^2$$

| Symbol | Meaning |
|--------|---------|
| $x_i$ | A data point |
| centroid | The cluster center |

---

## 3. Intuition

**Good clustering** → points are close to their centroid → **low inertia**

**Bad clustering** → points are far from their centroid → **high inertia**
```
Good cluster:         Bad cluster:
   • •               •         •
  • X •                   X
   • •                        •
```

---

## 4. Why Squared Distance?

Squaring penalizes large distances more heavily:

| Distance | Squared |
|----------|---------|
| 1 | 1 |
| 2 | 4 |
| 5 | 25 |

> Outliers have a strong effect on cluster assignment.

---

## 5. Accessing Inertia in Scikit-Learn
```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

print(kmeans.inertia_)  # e.g. 212.45
```

> **Lower inertia → better clustering.**

---

## 6. Important Insight — The Problem With Inertia

Increasing K **always** reduces inertia:

| K | Inertia |
|---|---------|
| 2 | 900 |
| 3 | 500 |
| 4 | 300 |
| 5 | 150 |

More clusters → points are always closer to a centroid. This raises an important question:

$$\text{How do we choose the right K?}$$

> The next topic solves this.

---

## 🔑 Key Summary

$$\text{Inertia} = \sum \| x_i - \text{centroid} \|^2$$

| | |
|--|--|
| **Goal** | Minimize inertia (WCSS) |
| **Low inertia** | Points are close to their centroids → good clustering |
| **High inertia** | Points are far from centroids → poor clustering |
| **Trap** | More clusters always lowers inertia → need a smarter way to choose K |

---

# Topic 4 — Choosing the Number of Clusters (Elbow Method)

## 📍 Where to Read
**9_Unsupervised Learning Techniques.pdf**
- Section discussing Choosing the Number of Clusters
- Figures showing the Elbow Curve

> One of the most important practical problems in clustering.

---

## 1. The Problem

K-Means requires us to choose K upfront — but the correct value is usually **unknown**.

$$\text{Are there 3 customer segments? 5? 10?}$$

We need a systematic way to decide.

---

## 2. The Elbow Method

1. Run K-Means for several values of K
2. Record the inertia for each K
3. Plot **K vs. Inertia** and look for the bend
```
Inertia
  |
1000|\
    | \
 600|  \
    |   \
 300|    \___________
    |
    +------------------
      1  2  3  4  5  6
              K
```

The bend in the curve is called the **elbow** — it suggests the best K.

---

## 3. Why This Works

| Region | What happens |
|--------|-------------|
| Before elbow | Adding clusters **greatly** reduces inertia |
| After elbow | Improvement becomes **small** — extra clusters add little value |

---

## 4. Example Interpretation

| K | Inertia |
|---|---------|
| 1 | 1000 |
| 2 | 600 |
| 3 | 330 |
| 4 | 250 |
| 5 | 220 |

Elbow at **K = 3** — improvements after that are small.

---

## 5. Code Example
```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=1000, centers=5, random_state=42)

inertias = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

plt.plot(range(1, 10), inertias, marker='o')
plt.xlabel("Number of clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

---

## 6. Limitations

Sometimes the elbow is **not clearly visible** — the curve is smooth with no obvious bend. In these cases use the **Silhouette Score** instead (covered in the next topic).

---

## 🔑 Key Summary

**Elbow Method — choose optimal K:**

1. Run K-Means for multiple values of K
2. Compute inertia for each
3. Plot K vs. inertia
4. Choose the **elbow point** — where improvement starts to flatten

$$\text{Elbow} = \text{point of diminishing returns in inertia reduction}$$

> **The Elbow Method** helps determine the optimal number of clusters (K)
> by finding the point where inertia stops decreasing rapidly.

---

# Topic 5 — Silhouette Score (Evaluating Clustering Quality)

## 📍 Where to Read
**9_Unsupervised Learning Techniques.pdf**
- Section: Silhouette Score

> The Elbow Method can be ambiguous — the Silhouette Score gives a more precise measure of clustering quality.

---

## 1. What is the Silhouette Score?

The Silhouette Score measures **how well each point fits within its own cluster compared to other clusters** — evaluating both compactness and separation.

---

## 2. Silhouette Value for a Single Point

For each data point, compute:

$$a = \text{average distance to points in the \textbf{same} cluster}$$
$$b = \text{average distance to points in the \textbf{nearest other} cluster}$$

$$\text{Silhouette Score} = \frac{b - a}{\max(a, b)}$$

---

## 3. Score Range

| Score | Meaning |
|-------|---------|
| **+1** | Perfect clustering — point is well inside its cluster |
| **0** | Clusters overlap |
| **-1** | Point is likely in the wrong cluster |

---

## 4. Choosing K with Silhouette Score

| K | Silhouette Score |
|---|-----------------|
| 2 | 0.41 |
| 3 | **0.58** ← best |
| 4 | 0.55 |
| 5 | 0.49 |

> Choose the K with the **highest silhouette score**.

---

## 5. Code Example
```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=500, centers=4, random_state=42)

kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

score = silhouette_score(X, labels)
print("Silhouette Score:", score)  # e.g. 0.63
```

---

## 6. Silhouette Plot

A silhouette plot shows how well each point fits its cluster:

| Plot Shape | Meaning |
|------------|---------|
| Wide positive bars | Good, compact clusters |
| Many values near 0 or negative | Poor separation or wrong assignments |

---

## 7. Elbow Method vs. Silhouette Score

| Method | Use When |
|--------|----------|
| **Elbow Method** | Quick first look at inertia drop-off |
| **Silhouette Score** | Elbow is unclear or you need a precise quality metric |

---

## 🔑 Key Summary

$$\text{Silhouette Score} = \frac{b - a}{\max(a, b)}$$

| Value | Interpretation |
|-------|---------------|
| Close to **+1** | Well-separated, compact clusters |
| Near **0** | Overlapping clusters |
| **Negative** | Points assigned to wrong clusters |

> **Higher silhouette score = better clustering.** Use it to compare K = 2, 3, 4, 5... and pick the best.

---

## 💡 Concept Clarification — Elbow Method vs. Silhouette Score

> *"If Silhouette Score can find the perfect number of clusters, why do we even need the Elbow Method?"*

---

### Short Answer

Silhouette Score is **not always reliable on its own** — both methods have weaknesses, so they are best used **together**.

---

### Why We Still Use the Elbow Method

**① Elbow Method is Faster**

Silhouette Score requires computing **pairwise distances** between all points:
$$\text{Complexity} \approx O(n^2)$$

Inertia is already computed **for free** during K-Means training — so the elbow plot costs nothing extra.

**② Silhouette Score Can Fail**

Silhouette assumes clusters are **convex and well-separated**. Real datasets often have:
- Irregular cluster shapes
- Varying densities
- Overlapping clusters

In these cases, silhouette may suggest the **wrong K**.

**③ Elbow Gives Intuition**

The elbow curve reveals *how much* clustering improves as K grows:

| K | Inertia |
|---|---------|
| 2 | 900 |
| 3 | 450 |
| 4 | 300 |
| 5 | 270 |
| 6 | 260 |

> Most improvement happens before K = 4 — the elbow makes this obvious at a glance.

---

### Best Practice — Use Both Together
```
Step 1 → try K = 2 to 10
Step 2 → plot elbow curve      (quick rough estimate)
Step 3 → compute silhouette scores  (confirm best K)
Step 4 → visualize clusters    (sanity check)
Step 5 → choose K that balances both signals
```

---

### Practical Rule

| Method | Role |
|--------|------|
| **Elbow Method** | Fast rough estimate of K |
| **Silhouette Score** | Precise validation of clustering quality |

> Neither method is perfect alone — combining them leads to more reliable decisions.

# Topic 6 — The K-Means Initialization Problem

## 📍 Where to Read
**9_Unsupervised Learning Techniques.pdf**
- Section: K-Means Initialization

> A very important practical issue with K-Means.

---

## 1. The Problem

K-Means starts by **randomly** placing centroids. Different random starts can produce very different results:
```
Run 1 → good clusters
Run 2 → poor clusters
Run 3 → different clusters entirely
```

---

## 2. Local Optimum Problem

K-Means minimizes inertia, but can get **stuck in a local optimum** — a solution that can't be improved by the algorithm, but isn't the best possible one either.

$$\text{Global minimum} = \text{best clustering}$$
$$\text{Local minimum} = \text{suboptimal clustering the algorithm gets stuck in}$$

---

## 3. Solution 1 — Multiple Initializations (`n_init`)

Run K-Means several times with different random starts, then keep the result with the **lowest inertia**:
```python
KMeans(n_clusters=4, n_init=10)
# Runs K-Means 10 times, keeps the best result
```

---

## 4. Solution 2 — K-Means++ Initialization

A smarter initialization that **spreads centroids far apart** from the start, producing better clustering and faster convergence.

**How it works:**

**Step 1** — Choose the first centroid randomly.

**Step 2** — Choose each subsequent centroid with probability proportional to:
$$\text{distance}^2 \text{ from existing centroids}$$

Points **far from current centroids** are more likely to be selected next.

**Step 3** — Repeat until K centroids are chosen.

---

## 5. Advantages of K-Means++

- More stable clustering results
- Faster convergence
- Avoids poor random initializations

> In Scikit-Learn, **K-Means++ is already the default**:
> ```python
> KMeans(init="k-means++")  # default behaviour
> ```

---

## 🔑 Key Summary

| Problem | Solution |
|---------|----------|
| Random init → different results each run | Use `n_init` to run multiple times |
| May converge to local optimum | Use **K-Means++** for smarter initialization |

**K-Means++ key idea:**
$$\text{Select initial centroids that are far apart} \rightarrow \text{better starting point} \rightarrow \text{better clustering}$$

> **K-Means++** improves clustering by selecting initial centroids that are far apart from each other,
> reducing the chance of poor local optima and improving convergence speed.

# Topic 7 — Mini-Batch K-Means (Scaling K-Means to Huge Datasets)

## 📍 Where to Read
**9_Unsupervised Learning Techniques.pdf**
- Section: Mini-Batch K-Means

> Used when datasets become too large for standard K-Means.

---

## 1. The Problem with Standard K-Means

Every iteration processes **all samples** — for a dataset of 10 million samples × 100 features, this means computing distances to every centroid for every point on every pass.

$$\text{Result: K-Means becomes very slow for large datasets}$$

---

## 2. The Idea — Use Small Random Batches

Instead of updating centroids on the full dataset, Mini-Batch K-Means updates on a **small random batch** each iteration:
```
Standard K-Means:    entire dataset → update centroids
Mini-Batch K-Means:  batch 1 → update centroids
                     batch 2 → update centroids
                     batch 3 → update centroids ...
```

> Think of it like **stochastic gradient descent for clustering** — estimating the solution from partial data.

---

## 3. Speed vs. Accuracy Trade-off

| Algorithm | Speed | Accuracy |
|-----------|-------|----------|
| **K-Means** | Slower | Slightly better |
| **Mini-Batch K-Means** | Much faster | Slightly worse |

> In practice the accuracy difference is often **very small**.

---

## 4. Implementation in Scikit-Learn
```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=10000, centers=5, random_state=42)

mbk = MiniBatchKMeans(
    n_clusters=5,
    batch_size=100,
    random_state=42
)

labels = mbk.fit_predict(X)
```

> `batch_size=100` means the algorithm processes **100 samples per centroid update**.

---

## 5. When to Use Mini-Batch K-Means

Use it when:
- Dataset is extremely large
- Memory is limited
- Training time matters

**Common use cases:** customer click logs, web activity, image embeddings, sensor streams.

---

## 🔑 Key Summary

| | |
|--|--|
| **Purpose** | Scale K-Means to large datasets |
| **Idea** | Update centroids using small random batches |
| **Gains** | Much faster, scalable, lower memory usage |
| **Cost** | Slightly less accurate clustering |
```python
MiniBatchKMeans(n_clusters=5, batch_size=100)
```

> **Mini-Batch K-Means** is a scalable version of K-Means that updates cluster
> centroids using small random batches instead of the entire dataset.

# Topic 8 — Using K-Means for Image Segmentation

## 📍 Where to Read
**9_Unsupervised Learning Techniques.pdf**
- Section: Image Segmentation Using K-Means

> A real-world application of clustering in computer vision.

---

## 1. What is Image Segmentation?

**Image segmentation** divides an image into meaningful regions by **grouping similar pixels** together.

**Example regions:** sky | tree | grass | road

---

## 2. How K-Means Segments Images

Every pixel is treated as a **3D data point** based on its color:

$$\text{pixel} = (R, \ G, \ B) \quad \text{e.g. } (120, \ 80, \ 200)$$

K-Means groups pixels into **K color clusters**, finds the K dominant colors, and replaces each pixel with its cluster's centroid color.

---

## 3. Effect of Different K Values

| K | Result |
|---|--------|
| 2 | Very coarse — only 2 colors |
| 5 | Moderate detail |
| 10 | More detailed regions |

---

## 4. Steps for Image Segmentation
```
Step 1 → load image
Step 2 → reshape into pixel dataset  (width × height, 3)
Step 3 → apply K-Means clustering
Step 4 → replace each pixel color with its centroid color
```

---

## 5. Code Example
```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.datasets import load_sample_image
import numpy as np

# Load image and reshape into pixel dataset
china = load_sample_image("china.jpg")
X = china.reshape(-1, 3)

# Cluster pixels into 5 color groups
kmeans = KMeans(n_clusters=5, random_state=42)
labels = kmeans.fit_predict(X)

# Replace each pixel with its cluster centroid color
segmented_img = kmeans.cluster_centers_[labels]
segmented_img = segmented_img.reshape(china.shape)

plt.imshow(segmented_img.astype(np.uint8))
plt.axis("off")
plt.show()
```

**What happens internally:**
```
each pixel → assigned to nearest color cluster
cluster centroid → becomes the new pixel color
```

**Example clusters for K = 5:**
```
Cluster 1 → sky blue
Cluster 2 → dark green
Cluster 3 → grass green
Cluster 4 → brown
Cluster 5 → gray
```

---

## 6. Real-World Applications

Image segmentation with K-Means is used in:
- Medical imaging
- Self-driving cars
- Satellite imagery
- Object detection

---

## 🔑 Key Summary

**Idea:** Treat each pixel as a data point with RGB values.

| Step | Action |
|------|--------|
| 1 | Convert image → pixel dataset of shape `(width × height, 3)` |
| 2 | Run K-Means with chosen K |
| 3 | Assign each pixel to its nearest color cluster |
| 4 | Replace pixel colors with cluster centroid colors |

$$\text{Result: image compressed into K distinct color regions}$$

> **In image segmentation using K-Means**, each pixel is treated as a feature vector
> containing RGB color values — e.g. `(120, 80, 200)` means Red=120, Green=80, Blue=200.
> The image becomes a dataset of shape `(number_of_pixels, 3)`.

# Topic 9 — DBSCAN (Density-Based Clustering)

## 📍 Where to Read
**9_Unsupervised Learning Techniques.pdf**
- Section: DBSCAN

> One of the most important clustering algorithms after K-Means — solves several of K-Means' major limitations.

---

## 1. The Problem with K-Means

K-Means struggles with:
- Must choose K beforehand
- Assumes **spherical** clusters
- Sensitive to **outliers**
- Fails on **irregular** cluster shapes (e.g. two moon-shaped clusters)

---

## 2. What is DBSCAN?

**DBSCAN** = Density-Based Spatial Clustering of Applications with Noise

$$\text{Dense regions} \rightarrow \text{clusters} \qquad \text{Sparse regions} \rightarrow \text{noise}$$

---

## 3. Key Parameters

| Parameter | Meaning | Example |
|-----------|---------|---------|
| **ε (epsilon)** | Maximum distance to consider two points neighbors | `eps=0.5` |
| **min_samples** | Minimum neighbors within ε to form a core point | `min_samples=5` |

---

## 4. Three Types of Points

| Type | Definition |
|------|------------|
| **Core Point** | Has ≥ `min_samples` neighbors within ε — inside a dense region |
| **Border Point** | Near a core point but has fewer than `min_samples` neighbors — on cluster boundary |
| **Noise Point** | Not close to any cluster — treated as an **outlier** |

---

## 5. Why DBSCAN is Powerful

✅ Automatically finds the **number of clusters** — no K needed

✅ Handles **irregular cluster shapes** (moons, spirals, arbitrary shapes)

✅ Detects **outliers** natively (label = `-1`)

---

## 6. Code Example
```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

X, y = make_moons(n_samples=300, noise=0.05)

dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="Set1")
plt.title("DBSCAN on Moon-Shaped Clusters")
plt.show()
```

> `labels == -1` → **noise points** (outliers)

---

## 7. K-Means vs. DBSCAN

| | K-Means | DBSCAN |
|--|---------|--------|
| Requires K | ✅ Yes | ❌ No |
| Handles irregular shapes | ❌ No | ✅ Yes |
| Detects outliers | ❌ No | ✅ Yes |
| Sensitive to density variation | ❌ No | ⚠️ Yes |

---

## 8. Limitations of DBSCAN

- Struggles when clusters have **very different densities**
- Performance degrades in **high dimensions**
- Choosing the right **ε** can be difficult

---

## 🔑 Key Summary

| Concept | Meaning |
|---------|---------|
| **ε (epsilon)** | Neighborhood radius |
| **min_samples** | Minimum points to form a cluster |
| **Core point** | Dense region interior |
| **Border point** | Cluster edge |
| **Noise point** | Outlier — label `-1` |

$$\text{Dense region} \xrightarrow{\text{DBSCAN}} \text{Cluster} \qquad \text{Isolated point} \xrightarrow{\text{DBSCAN}} \text{Noise}$$

> **In DBSCAN**, ε (epsilon) defines the maximum distance between two points
> for them to be considered neighbors — e.g. `ε = 0.3` means only points within distance 0.3 are treated as neighbors.

# Topic 10 — Gaussian Mixture Models (GMM)

## 📍 Where to Read
**9_Unsupervised Learning Techniques.pdf**
- Section: Gaussian Mixture Models

> A probabilistic alternative to K-Means — the last major algorithm of Chapter 9.

---

## 1. Limitation of K-Means

K-Means assumes clusters are **spherical, equal-sized, and separated**. Real datasets often have clusters that are elongated, rotated, overlapping, or differently sized — K-Means struggles with all of these.

---

## 2. Idea of Gaussian Mixture Models

GMM assumes data is generated from a **mixture of Gaussian distributions**:

$$\text{GMM} = \text{weighted combination of multiple Gaussian distributions}$$
```
Cluster 1 → Gaussian 1
Cluster 2 → Gaussian 2
Cluster 3 → Gaussian 3
```

Each Gaussian forms an **elliptical shape** in multiple dimensions, allowing clusters to be stretched, rotated, and overlapping.

---

## 3. Hard Clustering vs. Soft Clustering

| | K-Means | GMM |
|--|---------|-----|
| **Type** | Hard clustering | Soft clustering |
| **Assignment** | Each point belongs to **exactly one** cluster | Each point has a **probability** of belonging to each cluster |

**Example — GMM soft assignment:**
```
point → 70% cluster A
point → 30% cluster B
```

---

## 4. Parameters of Each Gaussian

| Parameter | Symbol | Meaning |
|-----------|--------|---------|
| Mean | $\mu$ | Cluster center |
| Covariance | $\Sigma$ | Cluster shape and orientation |
| Weight | $\pi$ | Cluster probability (how large the cluster is) |

---

## 5. How GMM Learns — Expectation-Maximization (EM)

GMM fits clusters using the **EM algorithm**, which alternates between two steps:

**E-Step (Expectation):** Compute the probability that each point belongs to each cluster.

**M-Step (Maximization):** Update each Gaussian's parameters — mean, covariance, and weight.

$$\text{Repeat E} \leftrightarrow \text{M until convergence}$$

---

## 6. Code Example
```python
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=300, centers=3, random_state=42)

gmm = GaussianMixture(n_components=3, random_state=42)
labels = gmm.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="Set2")
plt.title("Gaussian Mixture Model Clustering")
plt.show()
```

> `n_components` = number of Gaussian clusters (analogous to K in K-Means)

## 2. K-Means Clustering

Partitions a dataset into **K clusters**, each represented by a **centroid**.

**Algorithm:**
1. Choose K
2. Initialize centroids randomly
3. Assign each point to nearest centroid (Euclidean distance)
4. Update centroid = mean of cluster points
5. Repeat until convergence

---

## 3. K-Means Objective — Inertia

$$\text{Inertia} = \sum \| x_i - \text{centroid} \|^2$$

> Lower inertia = better clustering. Increasing K **always** decreases inertia.

---

## 4. Choosing Number of Clusters

**Elbow Method:** Plot K vs. inertia — choose the bend (elbow point).

**Silhouette Score:**
$$\text{Silhouette} = \frac{b - a}{\max(a, b)}$$

| Score | Meaning |
|-------|---------|
| +1 | Perfect clustering |
| 0 | Overlapping clusters |
| -1 | Incorrect clustering |

> **Best practice:** Use elbow for quick estimate, silhouette to validate.

---

## 5. K-Means Initialization Problem

Random initialization can lead to **local optima**.

| Solution | How |
|----------|-----|
| **Multiple runs** | `n_init=10` — keep lowest inertia result |
| **K-Means++** | Choose initial centroids far apart — default in Scikit-Learn |

## 8. DBSCAN

**Density-Based Spatial Clustering of Applications with Noise**

$$\text{Dense regions} \rightarrow \text{clusters} \qquad \text{Sparse regions} \rightarrow \text{noise}$$

| Parameter | Meaning |
|-----------|---------|
| **ε (epsilon)** | Max distance to consider two points neighbors |
| **min_samples** | Min points within ε to form a core point |

| Point Type | Definition |
|------------|------------|
| **Core** | ≥ `min_samples` neighbors within ε |
| **Border** | Near a core point, fewer neighbors |
| **Noise** | Isolated outlier — label = `-1` |

**Advantages:** No K needed, detects outliers, handles irregular shapes.

**Limitations:** Struggles with varying densities and high dimensions.

---

## 9. Gaussian Mixture Models (GMM)

Data is modeled as a **mixture of Gaussian distributions** — one per cluster.

| Parameter | Symbol | Meaning |
|-----------|--------|---------|
| Mean | $\mu$ | Cluster center |
| Covariance | $\Sigma$ | Cluster shape & orientation |
| Weight | $\pi$ | Cluster probability |

**Hard vs. Soft Clustering:**

| | K-Means | GMM |
|--|---------|-----|
| Assignment | Hard — one cluster only | Soft — probabilities per cluster |
| Cluster shape | Spherical | Elliptical ✅ |
| Overlapping | ❌ | ✅ |

**Learning algorithm — EM (Expectation-Maximization):**

$$\text{E-Step: estimate cluster probabilities} \leftrightarrow \text{M-Step: update } \mu, \Sigma, \pi$$

**Model selection:** Use **BIC** or **AIC** — lower = better.
