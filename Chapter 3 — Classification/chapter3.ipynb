## ðŸ”¹ 2ï¸âƒ£ Precision & Recall

### ðŸŽ¯ Aim

To measure quality of positive predictions.

**Accuracy answers:**  
"How often am I correct overall?"

**Precision answers:**  
"When I say positive, how often am I right?"

**Recall answers:**  
"Of all real positives, how many did I catch?"

### ðŸ§  Intuition

Think medical diagnosis:

* **Precision** = Avoid false alarms.
* **Recall** = Don't miss sick patients.

High precision â†’ cautious model.  
High recall â†’ aggressive model.

**Tradeoff is inevitable.**

### Then comes:

**F1 Score**

Harmonic mean of precision and recall.

Why harmonic mean?  
Because it punishes extreme imbalance between them.

---

## ðŸ”¹ 3ï¸âƒ£ Precision / Recall Tradeoff

### ðŸŽ¯ Aim

To understand that classifiers output scores, not just labels.

Most classifiers compute:
* Probability
* Or decision score

Then apply a threshold (usually 0.5 or 0).

**Changing threshold shifts precision and recall.**

### ðŸ§  Intuition

Imagine a spam filter with threshold:

* Very low threshold â†’ catches everything (high recall, low precision)
* Very high threshold â†’ only super-certain spam (high precision, low recall)

**We are not training a different model.**  
We are just sliding a decision boundary.

### This is critical mastery concept:

**Performance tuning often means threshold tuning, not model retraining.**

---

## ðŸ”¹ 4ï¸âƒ£ ROC Curves

### ðŸŽ¯ Aim

To evaluate classifier performance across all thresholds.

**ROC plots:**
* True Positive Rate (Recall)
* False Positive Rate

Instead of precision vs recall.

### ðŸ§  Intuition

Precision depends on class imbalance.  
ROC is insensitive to imbalance.

**ROC curve shows:**  
How separable the classes are.

**AUC (Area Under Curve):**
* 1.0 = perfect separation
* 0.5 = random guessing

### Important conceptual difference:

* **Precision/Recall curve** â†’ useful for imbalanced data.
* **ROC curve** â†’ useful for balanced comparison.

## ðŸŽ¯ Why Accuracy Is Misleading (Deeper Understanding)

Accuracy measures:

$$\frac{TP + TN}{Total}$$

### The problem is:

**Accuracy treats all errors equally.**

But in skewed datasets:
* The majority class dominates TN.
* You can get huge TN numbers just by predicting the majority class.

---

### If your model predicts:

### Then:
* TN â‰ˆ 90%
* TP = 0
* FN = 10%
* FP = 0

**Accuracy â‰ˆ 90%**

### But the model:
* Never detects a single 5.
* Completely useless for the task.

---

## ðŸ”¥ The Real Issue

**Accuracy ignores:**
* The cost of missing positives.
* The cost of false alarms.
* The imbalance of class distribution.

**Accuracy answers:**  
"How often am I correct?"

**But classification often requires:**  
"How good am I at detecting the important class?"

## Next Major Step: Confusion Matrix via Cross-Validated Predictions

### ðŸŽ¯ Aim

Get an honest set of predictions for the training set without cheating.

If we simply do:
```python
sgd_clf.predict(X_train)
```

we'd be evaluating on the same data the model learned from, which can be overly optimistic.

So we use **cross-validation predictions:**

**For each training instance, predict it using a model that did not train on that instance.**

That gives us predictions we can trust enough to build a confusion matrix and later precision/recall.

---

## ðŸ§  Intuition

`cross_val_predict` is like running 3 mini-experiments:
* Split training data into 3 folds
* Train on 2 folds, predict on the held-out fold
* Repeat until every instance has a "held-out" prediction

This is crucial because Chapter 3 is obsessed with a single theme:

**Evaluation must reflect real generalization, not memorization.**

---

## âš ï¸ What could go wrong

* It can take a bit of time (it trains 3 models here).
* If you accidentally evaluate on training predictions (non-CV), metrics can look "too good."
* If you shuffle/stratify incorrectly in other datasets, you can introduce leakage (MNIST here is fine for our learning setup).

## ðŸ”¥ What Is ROC?

**ROC = Receiver Operating Characteristic**

It is a curve that plots:

$$\textbf{True Positive Rate (TPR)} \quad \text{vs} \quad \textbf{False Positive Rate (FPR)}$$

as the threshold moves.

---

## ðŸŽ¯ Definitions (Precise)

### 1ï¸âƒ£ True Positive Rate (TPR)

This is just:

$$\text{TPR} = \text{Recall} = \frac{TP}{TP + FN}$$

It measures:

**Of all real 5s, how many did we catch?**

### 2ï¸âƒ£ False Positive Rate (FPR)

$$\text{FPR} = \frac{FP}{FP + TN}$$

It measures:

**Of all real NOT-5s, how many did we incorrectly call 5?**

This is different from precision.

**Precision asks:**
* When I say 5, how often am I right?

**FPR asks:**
* Among real negatives, how often do I create false alarms?

---

## ðŸ§  Intuition of ROC

Imagine sliding threshold from:
```
+âˆž â†’ -âˆž
```

**At extreme high threshold:**
* Predict nothing positive
* TPR = 0
* FPR = 0
* ROC starts at (0,0)

**At extreme low threshold:**
* Predict everything positive
* TPR = 1
* FPR = 1
* ROC ends at (1,1)

Between those extremes, we trace a curve.

---

## ðŸ”¥ What Does ROC Show Conceptually?

**ROC shows:**

**How well the model separates classes.**

**If classes are perfectly separable:**
* Curve goes straight up to (0,1)
* Then across to (1,1)

**If model is random guessing:**
* ROC is diagonal line
* TPR â‰ˆ FPR
* AUC = 0.5

---

## ðŸŽ¯ Why Is ROC Different from Precision-Recall?

Key difference:

|  | ROC | Precision-Recall |
|---|---|---|
|  | Uses TPR & FPR | Uses Precision & Recall |
|  | Insensitive to class imbalance | Sensitive to imbalance |
|  | Good for comparing models globally | Better for rare positives |

---

## ðŸ§  Deep Mastery Insight

**Precision depends on:**

$$\frac{TP}{TP+FP}$$

So if negatives dominate (like MNIST: 90% not-5),  
precision is highly influenced by class distribution.

**FPR divides by:**

$$FP + TN$$

which is total negatives.

So ROC stays stable even if class ratio changes.

## 2ï¸âƒ£ Binary Classification (5 vs Not-5)

### Setup
* Convert MNIST to binary: `y == "5"`

### Core Insight
Most classifiers output a **score**:

$$\text{score}(x)$$

Prediction is:

$$\text{score}(x) > \text{threshold}$$

> The model produces **scores**, not labels. Threshold turns score into decision. This distinction is fundamental.

---

## 3ï¸âƒ£ Confusion Matrix (Foundation of Everything)

|  | Pred + | Pred âˆ’ |
|---|---|---|
| **True +** | TP | FN |
| **True âˆ’** | FP | TN |

Everything comes from these four numbers.

### Deep Insight
Experts read confusion matrices **before** computing metrics. It tells you:
* What kind of mistakes dominate
* Whether the model is conservative or aggressive

---

## 4ï¸âƒ£ Accuracy Is Misleading

$$\text{Accuracy} = \frac{TP + TN}{\text{Total}}$$

**Problem:** Treats all errors equally. Fails in imbalanced datasets.

> Example: If 90% are not-5, predicting "not-5" always â†’ 90% accuracy. Completely useless model.

---

## 5ï¸âƒ£ Precision, Recall, F1

### Precision
$$\frac{TP}{TP + FP}$$
*"When I predict positive, how often am I right?"* â€” Measures quality of positive predictions.

### Recall (Sensitivity / TPR)
$$\frac{TP}{TP + FN}$$
*"How many real positives did I catch?"* â€” Measures coverage.

### Tradeoff
* Increasing threshold â†’ â†‘ Precision, â†“ Recall
* Decreasing threshold â†’ â†‘ Recall, â†“ Precision

> This is **not** model retraining. It's **policy adjustment**.

### F1 Score
$$2 \cdot \frac{Precision \cdot Recall}{Precision + Recall}$$

Harmonic mean. Punishes imbalance between precision and recall.

---

## 6ï¸âƒ£ Threshold Tuning

The classifier gives **scores**. You choose threshold based on:
* Cost of FN vs FP
* Business constraints
* Desired operating point

> There is no universally "correct" threshold. Machine learning becomes **decision theory** here.

---

## 7ï¸âƒ£ Precisionâ€“Recall Curve

Shows precision vs recall across **all thresholds**. Important in imbalanced datasets.

At extreme thresholds:
* Very high threshold â†’ Recall = 0
* Very low threshold â†’ Recall = 1, Precision â‰ˆ base rate

> Precision at recall=1 â‰ˆ prevalence of positive class. **Critical rare-event insight.**

---

## 8ï¸âƒ£ ROC Curve

Plots **TPR (Recall)** vs **FPR**:

$$FPR = \frac{FP}{FP + TN}$$

### ROC AUC
> AUC = probability that a random positive **ranks above** a random negative.

* `1.0` â†’ perfect separation
* `0.5` â†’ random guessing

### ROC vs PR (Critical Distinction)

| | ROC | PR |
|---|---|---|
| Class imbalance | Insensitive | Sensitive |
| Measures | Ranking ability | Usefulness for detecting positives |
| Best for | General use | Rare-event problems |

---

## 9ï¸âƒ£ Multiclass Classification

Binary â†’ one score. Multiclass â†’ **multiple scores**.

For each class $k$: $\text{score}_k(x)$

Prediction:
$$\arg\max_k \text{ score}_k(x)$$

### One-vs-Rest (OvR)
* Train one binary classifier per class
* Choose highest score
* `SGDClassifier` does this internally
